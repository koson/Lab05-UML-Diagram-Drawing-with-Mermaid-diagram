# Lab 5.3 การวาด sequence diagram ด้วย mermaid diagram

### 5.3.1 องค์ประกอบของ sequence diagram

* Class หรือ Object
  - เขียนแทนด้วย รูปสี่เหลี่ยม
  - ภายในกรอบมีชื่อของ object/class ในรูปแบบ {Object}:Class
  - {Object} หมายถึง การระบุหรือไม่ระบุ object ก็ได้

* เส้นแสดงลำดับเวลา

    - แทนด้วย เส้นตรงประแนวตั้ง
    - เวลาจะเดินจากด้านบนมาสู่ด้านล่าง
    - กิจกรรมบนสุดถือเป็นกิจกรรมแรก และกิจกรรมที่อยู่ต่ำลงมาจะเป็นกิจกรรมที่เกิดขึ้นต่อจากนั้น

* เส้นแสดงกิจกรรม

    - แทนด้วย ลูกศรแนวนอน
    - ชี้จาก class/object หนึ่งไปยัง class/object ตัวต่อไป
    - ชื่อของกิจกรรมจะต้องเป็น Function ที่มีอยู่ใน
    - class/object ที่ลูกศรชี้ไป
    - ชื่อจะอยู่ในรูปแบบ {[Condition]}Function

### 5.3.3 รูปแบบของลูกศรใน sequence diagram


| Type	| Description|
|:-----:| -----------|
|->	    |Solid line without arrow  |
|-->	|Dotted line without arrow|
|->>	|Solid line with arrowhead|
|-->>	|Dotted line with arrowhead|
|<<->>	|Solid line with bidirectional arrowheads (v11.0.0+)|
|<<-->>	|Dotted line with bidirectional arrowheads (v11.0.0+)|
|-x	    |Solid line with a cross at the end|
|--x	|Dotted line with a cross at the end|
|-)	    |Solid line with an open arrow at the end (async)|
|--)	|Dotted line with a open arrow at the end (async)|

## 2. เขียน code ในภาษา mermaid diagram เพื่อวาดตามไดอะแกรมนั้น

````
sequenceDiagram
    participant Student
    participant WebSite as Web Site
    participant WebServer as Web Server
    participant Admin
    participant DB as DB

    Student ->> WebSite: 1. Click on register tab()
    WebSite ->> WebServer: 2. Request register page()
    WebServer ->> WebSite: 3. Return register page()
    WebSite ->> Student: 4. Display Register field()
    Student ->> WebSite: 5. Enter information()
    Student ->> WebSite: 6. Click submit button()
    WebSite ->> WebServer: 7. Send new student information()
    WebServer ->> Admin: 8. Verify new student information()
    Admin ->> WebServer: 9. Invalid new student()
    WebServer ->> WebSite: 10. Send errors()
    Student ->> WebSite: 11. ReEnter information()
    Admin ->> WebServer: 12. Valid new student()
    WebServer ->> DB: 13. Save()
    WebServer ->> Student: 14. Send acceptance email()
````

````
sequenceDiagram
    participant OrderInput as Order Input Window
    participant Order as Order
    participant OrderEntry1 as Order Entry
    participant OrderEntry2 as Order Entry
    participant Reorder as Reorder
    participant Delivery as Delivery

    OrderInput ->> Order: Prepare()
    activate Order
    Order ->> OrderEntry1: *Prepare()
    activate OrderEntry1
    OrderEntry1 ->> OrderEntry1: check()
    alt [check == TRUE]
        OrderEntry1 ->> OrderEntry1: Delete()
    end
    OrderEntry1 ->> OrderEntry2: isReorderNeeded()
    alt [isReorderNeeded == TRUE]
        OrderEntry2 ->> Reorder: New()
        activate Reorder
        Reorder ->> Delivery: Deliver()
        deactivate Reorder
    end
    deactivate OrderEntry1
    deactivate Order
````

````
sequenceDiagram
    participant Customer
    participant ApplicationInterface as Application interface
    participant ItemId as Item id
    participant PurchaseInterface as Purchase interface
    participant Checkout
    participant NormalOrder as Normal order

    Customer ->> ApplicationInterface: browse item
    Customer ->> ApplicationInterface: select item
    ApplicationInterface ->> ItemId: query item
    ItemId ->> ApplicationInterface: get selected item
    ApplicationInterface ->> ItemId: purchase
    ApplicationInterface ->> PurchaseInterface: purchase
    PurchaseInterface ->> Checkout: authorizepurchase
    Checkout ->> NormalOrder: authorization
    NormalOrder -->> Checkout: authorization
    Checkout -->> PurchaseInterface: authorizepurchase
    ApplicationInterface -->> Customer: display selected item
    ApplicationInterface -->> Customer: displaypriceinfo
````
